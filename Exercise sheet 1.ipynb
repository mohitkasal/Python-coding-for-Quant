Exercise 1: Find the sum of numbers from 1 to n

```python
def sum_n(n):
    total = 0
    for i in range(1 , n+1):
        total = total + i
    return total
print(sum_n(100))
```

Shorter Version 

```python
def sum_n(n):
    return sum(range(1,n+1))
print(sum_n(5))    
```

Exercise 2: Print all the even numbers from 1 to 50

```python
for i in range(1,51):
    if i % 2 != 0:
        print(i)        
```

Sum of all the above odd numbers: 

```python
import numpy as np

x=[0]

for i in range(1,51):
    if i % 2 != 0:
        x.append(x[-1]+i)
print(x[-1])        
```

More compact way: (Even numbers)

```python
total = 0
for i in range(2,51,2):
    total+=i

print(total)    
```

Exercise 3: Max, min, sort a list

```python
arr = [1,5,2,9,3]
print(max(arr))
```

```python
arr = [1, 45, -1, 2]
print(min(arr))
```

```python
arr = [12,5,6,4,1,8]
print(sorted(arr))
```

Two common ways of doing sorting:

```python
arr = [1,5,3,0,7]
new_arr = sorted(arr)
print(new_arr)
print(arr)
```

```python
arr = [1,5,2,9,3]
arr.sort()
print(arr)
```

Exercise 4: Create a list of cubes from 1 to 10

```python
squares = []

for i in range(1,11):
    squares.append(i**3)

print(squares)    
```

Shorter version: Also called the list comprehension

```python
squares = [i**3 for i in range(1,11)]
print(squares)
```

Exercise 5: Vector operations with NumPy

```python
import numpy as np 

x = np.array([1,0,0])
y = np.array([0,1,0])
z = np.array([0,0,1])

print(x+y+z)

print(x*y*z)

print(np.dot(x,np.cross(y,z)))
```

Exercise 6: Generate 1000 normal random numbers and compute mean and variance

```python
import numpy as np

data = np.random.normal(5,2,1000)

print("Mean = ", np.mean(data))
print("Variance = ", np.var(data))
```
