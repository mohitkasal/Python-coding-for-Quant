Medical test

```python
import numpy as np

N = 100000

disease = np.random.rand(N) < 0.01

test_pos = np.where(
    disease, 
    np.random.rand(N) < 0.95,
    np.random.rand(N) < 0.05
)

prob = np.sum(disease & test_pos) / np.sum(test_pos)

print(prob)
```

Coin guessing

```python
import numpy as np 

N = 10000

#0 is fair and 1 is biased
coin = np.random.choice([0,1], size = N)

#toss
result = np.where(
    coin == 0,
    np.random.rand(N) < 0.5,
    np.random.rand(N) < 0.9
)

#conditional probability
mask = result == 1 #heads observed 

prob = np.mean(coin[mask] == 1)

print(prob)
```

Conditional probability in time series

```python
import numpy as np 

N = 10000

r = np.random.normal(0,1, N)

#shift data
r_t = r[: -1]
r_next = r[1:]

mask = r_t > 0

x = r_next[mask] > 0
prob = np.mean(r_next[mask] > 0)

print("Total points:", len(r))
print("Points where r_t > 0:", np.sum(mask))
print("Points where r_t+1 > 0:", np.sum(x))
print("P(r_t+1 > 0 | r_t > 0 ):", prob)
```

Introduce dependence

```python
import numpy as np 

N = 10000

r = np.zeros(N)

for t in range(1,N):
    r[t] = 0.5 * r[t-1] + np.random.normal(0,1)

r_t = r[:-1]
r_next = r[1:]

mask = r_t > 0

prob = np.mean(r_next[mask] > 0)  

print(prob)
```

Trading Signal

```python
import numpy as np 

N = 10000

r = np.zeros(N)

for t in range(1,N):
    r[t] = 0.5 * r[t-1] + np.random.normal(0,1)

r_t = r[:-1]
r_next = r[1:]

prediction = r_t > 0 

actual = r_next > 0 

accuracy = np.mean(prediction == actual)

print(accuracy)
```
