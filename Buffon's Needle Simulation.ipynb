```python
import numpy as np
import matplotlib.pyplot as plt

#parameters
L = 1.0   #needle length
d = 1.0   #distance between lines 
N = 100000

#generate random variables
theta = np.random.uniform(0, np.pi/2, N)
x = np.random.uniform(0, d/2, N)

#crossing condition
cross = x <= (L/2)*np.sin(theta)

#probability
prob = np.mean(cross)

print("Estimated probability = ", prob)

#theoretical value
theory = 2*L / (np.pi * d)
print("Theoretical probability = ", theory)

#plot
plt.scatter(theta[:1000], x[:1000], alpha=0.3)
plt.xlabel("theta")
plt.ylabel("x")
plt.title("Samples from Joint PDF")
plt.show()
```
