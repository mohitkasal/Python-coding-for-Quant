Estimate Probability

```python
import numpy as np 

N=10000
success = 0

for x in range(N):
    tosses = np.random.choice([0,1], size = 3)

    if np.sum(tosses) >= 1:
        success +=1

prob = success/N
print("Estimated probability = ", prob)
```

Law of large number

```python
import numpy as np 
import matplotlib.pyplot as plt

N = 1000
tosses = np.random.choice([0,1],size = N)

running_average = []

for i in range(1,N+1):
    running_average.append(np.mean(tosses[:i]))

plt.plot(running_average)
plt.xlabel("Number of tosses")
plt.ylabel("Estimated P(Heads)")
plt.show()
```

Random Walk

```python
import numpy as np 
import matplotlib.pyplot as plt

steps = 200
position = 0
positions = [position]

for x in range(steps): 
    step = np.random.choice([1,-1])
    position += step
    positions.append(position)

plt.plot(positions)
plt.xlabel("Step")
plt.ylabel("Position")
plt.title("Random Walk")
plt.show()
```

Many random Walks

```python
import numpy as np 
import matplotlib.pyplot as plt

steps = 200

for _ in range(1000):
    position = 0
    positions = [position]
    for _ in range(steps):
        step = np.random.choice([-1,1])
        position += step
        positions.append(position)
    plt.plot(positions)

plt.show()
```
