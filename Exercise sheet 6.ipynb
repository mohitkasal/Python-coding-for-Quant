Binomial probability

```python
import math

n = 5
k = 3
p = 0.5

prob = math.comb(n,k)*(p**k)*((1-p)**(n-k))

print(prob)
```

Same problem as before, but now we use SciPy

```python
from scipy.stats import binom

n = 5
p = 0.5

prob = binom.pmf(3,n,p)

print(prob)
```

```python
from scipy.stats import binom

n = 5
p = 0.5

prob = binom.cdf(3,n,p)

print(prob)
```

PDF of Normal Distribution

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

print(norm.pdf(1, loc = 0, scale = 1))

x = np.linspace(-3, 3, 100)
y = norm.pdf(x)

z = plt.plot(x,y)
plt.show(z)
```

Probability using CDF
