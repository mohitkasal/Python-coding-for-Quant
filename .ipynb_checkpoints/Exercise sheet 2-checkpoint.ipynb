Sum of a list

```python
arr = [2,4,6,8]

#using loop 
total = 0 
for x in arr:
    total = total+x
print(total)  

#using built in sum 
print(sum(arr))
```

Count even numbers

```python
arr = [1,2,3,4,5,6]

count = 0 
for x in arr: 
    if x % 2 == 0:
        count += 1

print(count)        
```

Find largest number manually

```python
arr = [3,7,2,9,5]

largest = arr[0]

for x in arr:
    if x > largest: 
        largest = x

print(largest)        
```

Reverse a list

```python
arr = [1,2,3,4]

reversed_arr = arr[::-1]

print(reversed_arr)
```

```python
import numpy as np

arr = np.array([10,20,30,40])
print(np.mean(arr))
```

10 Coin tosses

```python
import numpy as np 

toss = np.random.choice(["Heads","Tails"], size = 10)

print(toss)
```

Count heads

```python
import numpy as np 

tosses = np.random.choice(["Heads","Tails"], size = 10)

count_heads = 0
for toss in tosses: 
    if toss == "Heads":
       count_heads += 1

print(tosses)
print(count_heads)
```

Estimate probability of heads

```python
import numpy as np 

tosses = np.random.choice([0,1], size = 100)

print("Estimated P(Heads) = ", np.mean(tosses))
```
