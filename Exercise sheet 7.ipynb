Monte Carlo simulation of a probability problem where we have 10 members on our subscription based service. The probability that one of the members churns per month is 0.3. What is the probability that atleast one of the members doesn't churn in 12 months.

Stupid way of doing it: 

```python
import numpy as np 

N = 100000

num_members = 10
months = 12
p_survive_month = 0.7

success = 0

for _ in range(N): 
    survived_members = 0

    for _ in range(num_members):
        #simulate 12 months for one member
        months_survived = np.random.rand(months) < p_survive_month

        if np.all(months_survived):     #survived all 12 months
            survived_members += 1
    if survived_members >=1:
        success += 1

prob = success/N
print("Estimated probability = ", prob)
```

Better way of solving the problem: 

```python
import numpy as np 

N = 100000 

# shape: (simulations, members, months)
data = np.random.rand(N,10,12) < 0.7

#check survival across all the months
survived = np.all(data, axis = 2)

#check atleast one member survives
at_least_one = np.any(survived, axis = 1)

prob = np.mean(at_least_one)

print("Estimated probability = ", prob)
```
